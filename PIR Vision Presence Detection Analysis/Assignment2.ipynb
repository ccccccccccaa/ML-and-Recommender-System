### KNN modeling

```python
import numpy as np
# Create features (X) by dropping the first two columns and the target column
# as date and time is not needed for data training.
pir_cols = [f"PIR_{i}" for i in range(1, 56)]
#using .values to convert this into an array with numpy
X = pirvision[pir_cols + ['Temperature_F']].values
y = combine['Label'].values #assign Label as target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
#Kfold validation
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=4)

for fold, (train_index, test_index) in enumerate(kf.split(X)):

    X_train_fold, X_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    clf.fit(X_train_fold, y_train_fold)
    score = clf.score(X_test_fold, y_test_fold)
    print(f"  Fold {fold} accuracy: {score:.4f}")
```

```python
#Tuning the n_neighbor values to find the optimized value
```

```python
for k in range(1,21):#from 1 to 20, find the best K value for KNN
    clf = KNeighborsClassifier(n_neighbors=k)
    fit = clf.fit(X_train, y_train)
    y_pre = fit.predict (X_test)
    #use .score to print out the accuracy for each k value
    print(f"k={k} Test Accuracy: {clf.score(X_test, y_test):.5f}")
    cm = confusion_matrix(y_test,y_pre)
    print(cm)
```

```python
# when n_neighbors = 3, the accuracy is the highest = 0.9670
```

```python
for k in range(1,21):#from 1 to 20, find the best K value for KNN
    clf = KNeighborsClassifier(n_neighbors=k, weights='distance')
    fit = clf.fit(X_train, y_train)
    y_pre = fit.predict (X_test)
    #use .score to print out the accuracy for each k value
    print(f"k={k} Test Accuracy: {clf.score(X_test, y_test):.5f}")
    cm = confusion_matrix(y_test,y_pre)
    print(cm)
```

```python
#when the parameter weights is set to 'distance',
#the the accuracy maxes at 0.96766 with K=9
```

```python
for k in range(1,21):#from 1 to 20, find the best K value for KNN
    clf = KNeighborsClassifier(n_neighbors=k, weights='distance', p=1)
    fit = clf.fit(X_train, y_train)
    y_pre = fit.predict (X_test)
    #use .score to print out the accuracy for each k value
    print(f"k={k} Test Accuracy: {clf.score(X_test, y_test):.5f}")
    cm = confusion_matrix(y_test,y_pre)
    print(cm)
```

```python
clf = KNeighborsClassifier(n_neighbors=9, weights='distance', p=2)
fit = clf.fit(X_train, y_train)
y_pre = fit.predict (X_test)
cm = confusion_matrix(y_test,y_pre)

print(cm)
print('Accuracy:', clf.score(X_test, y_test))
print(classification_report(y_test,y_pre))

kf = KFold(n_splits=5, shuffle=True, random_state=4)
for fold, (train_index, test_index) in enumerate(kf.split(X)):

    X_train_fold, X_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    score = clf.score(X_test_fold, y_test_fold)
    print(f"  Fold {fold} accuracy: {score:.4f}")
```

#### now, with p=1(manhattan_distance), the model doesn't have a new peak again in terms of accuracy compared with the previous tests
#### Therefore, the best parameters for this model tested on test data set is

#### KNeighborsClassifier(n_neighbors=9, weights='distance')
